```python
#Import libraries ----
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load dataset ----
data = pd.read_csv("/content/SampleSuperstore.csv")
```

```python
#Inspect data ----
print("Column Names:", data.columns)
print(data.head())
```

```python
#Clean data ----
data.columns = data.columns.str.strip()
data.drop_duplicates(inplace=True)
data.dropna(inplace=True)
```

```python
# Basic stats ----
print("\nTotal Sales:", data['Sales'].sum())
print("Total Profit:", data['Profit'].sum())
```

```python
# Sales by Region ----
region_sales = data.groupby('Region')['Sales'].sum().reset_index()
sns.barplot(x='Region', y='Sales', data=region_sales)
plt.title("Sales by Region")
plt.ylabel("Total Sales")
plt.show()
```

```python

# Profit by Category ----
category_profit = data.groupby('Category')['Profit'].sum().reset_index()
sns.barplot(x='Category', y='Profit', data=category_profit)
plt.title("Profit by Category")
plt.ylabel("Total Profit")
plt.show()
```

```python
# Top 10 Profitable Sub-Categories ----
top_products = data.groupby('Sub-Category')['Profit'].sum().sort_values(ascending=False).head(10)
top_products.plot(kind='bar', figsize=(8,4), title="Top 10 Profitable Sub-Categories")
plt.ylabel("Total Profit")
plt.show()
```

```python

# Discount vs Profit ----
sns.scatterplot(x='Discount', y='Profit', data=data)
plt.title("Impact of Discount on Profit")
plt.show()
```

```python
# Sales vs Quantity ----
sns.scatterplot(x='Quantity', y='Sales', data=data)
plt.title("Sales vs Quantity")
plt.show()
```
